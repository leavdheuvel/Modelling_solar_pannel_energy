```python
import pandas as pd
import numpy as np
```

```python
ts = pd.read_csv('etmgeg_269.txt', index_col = 'YYYYMMDD')
ts.head()
```

```python
ts.columns
```

```python
# Strip de spaces
ts = ts.rename(columns=lambda x: x.strip())
ts.columns
```

```python
cols = ['TG', 'SQ', 'DR', 'RH']
```

```python
ts = ts[cols]
ts.info()
```

```python
# Functie om tabs/spaties te verwijderen en NaN invullen
def remove_space(df):
    # remove front and ending blank spaces
    df = df.replace({"^\s*|\s*$":""}, regex=True) 
    
    # if there remained only empty string "", change to Nan
    df = df.replace({"":np.nan}) 
    return df
```

```python
# Maak NaN van spaces
ts = remove_space(ts)
ts.dropna(inplace = True)
ts.info()
```

```python
# Date-time index
ts.index = pd.to_datetime(ts.index, format = '%Y%m%d')
ts.head()
```

```python
ts.to_csv('Weer_data.csv')
```
